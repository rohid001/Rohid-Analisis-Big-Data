```python
#Plotting bobot pada assessments
# Group by 'final_result' and calculate the sum of 'studied_credits' for each group
result_weight = df_assessments.groupby('assessment_type')['weight'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(result_weight, labels=result_weight.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Plotting pada tipe Assessment beserta Bobotnya')
plt.show()
print("TMA memiliki bobot dengan kesulurhan sebesar 42,1%")
print("CMA memiliki bobot dengan kesulurhan sebesar 2,1%")
print("EXAM memiliki bobot dengan kesulurhan sebesar 55,8%")
```
