```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_studentRegis = pd.read_csv('Datasets/studentRegistration.csv')
```

```python
# Menampilkan 5 baris pertama pada student registration
df_studentRegis.head()
```

```python
# Menampilkan 5 baris terakhir pada student registration
df_studentRegis.tail()
```

```python
print("Count missing value: \n")
df_studentRegis.isnull().sum()
```

```python
# Convert 'date_registration' and 'date_unregistration' to numeric, handling errors
df_studentRegis['date_registration'] = pd.to_numeric(df_studentRegis['date_registration'], errors='coerce')
df_studentRegis['date_unregistration'] = pd.to_numeric(df_studentRegis['date_unregistration'], errors='coerce')

# Plotting
plt.figure(figsize=(10, 6))

# Plot date_registration
plt.scatter(df_studentRegis['id_student'], df_studentRegis['date_registration'], label='Date Registration', color='blue', alpha=0.6)

# Plot date_unregistration
plt.scatter(df_studentRegis['id_student'], df_studentRegis['date_unregistration'], label='Date Unregistration', color='red', alpha=0.6)

plt.xlabel('Student ID')
plt.ylabel('Days Relative to Start')
plt.title('Date Registration and Unregistration')
plt.legend()
plt.show()
```
