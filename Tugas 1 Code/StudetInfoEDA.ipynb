```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ploting jumlah gender pada student info
gender_count = df_studentInfo['gender'].value_counts()

plt.bar(gender_count.index, gender_count.values)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribusi dari Gender student')
plt.show()
print("Pada bar chart di atas menunjukkan bahwa murid laki-laki lebih banyak daripada perempuan")
```

```python
# Ploting persebaran dari daerah pada student info
region_count = df_studentInfo['region'].value_counts()

plt.barh(region_count.index, region_count.values)
plt.xlabel('Count')
plt.ylabel('Region')
plt.title('Distribusi dari Region student')
plt.show()
```

```python
# Ploting tingkat pendidikan pada student info
region_count = df_studentInfo['highest_education'].value_counts()

plt.barh(region_count.index, region_count.values)
plt.xlabel('Count')
plt.ylabel('highest education')
plt.title('Distribusi dari Tingkat Pendidikan')
plt.show()
```

```python
# Kelompokkan berdasarkan 'final_result' dan hitung jumlah 'studied_credits' untuk setiap kelompok
result_credits = df_studentInfo.groupby('final_result')['studied_credits'].sum()

# plotting dengan pie chart
plt.figure(figsize=(8, 8))
plt.pie(result_credits, labels=result_credits.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Distribusi Kredit yang Dipelajari Berdasarkan Hasil Akhir')
plt.show()
```
